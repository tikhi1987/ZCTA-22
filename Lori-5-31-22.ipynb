1. The data for law enforcement agencies are provided per ZCTA (Zip Code Tabulation Area, which are often used rather than Zip codes) for each year (2003 through 2017). Would you be able to create a dataset that includes the average for all variables across all years (rather than for each year) for each ZCTA?

2. In the NanDa data, 922 = all law enforcement organizations. In terms of the types of law enforcement agencies that comprise “all law enforcement organizations", there are courts (code is 922110); police departments (code is 922120); lawyers and attorneys, specifically those that provide legal services for the government, such as prosecutors and public defenders (code is 922130); other lawyers (code is 541110); prisons, jails, and other correctional facilities (code is 922140); fire departments (code is 922160); and other organizations supporting law enforcement agencies, such as government emergency planning offices, public safety bureaus, and consumer product safety commissions (code is 922190). I am interested in getting the average for each kind of law enforcement agency as well as the average for “all law enforcement organizations” (total for all kinds of agencies).

```python
#Importing packages
import pandas as pd
import numpy as np
```

```python
#Importing law enforcement csv file from local drive as a pandas dataframe 
law_enforcement_df = pd.read_csv('nanda_lawenf_zcta_2003-2017_01P.csv')
law_enforcement_df.head()
```

```python
#grouping by zcta19 and taking the average
law_enforcement_df_group_zcta19 = law_enforcement_df.groupby(['zcta19'],as_index=False).mean()

#dropping "year" variable
law_enforcement_df_group_zcta19 = law_enforcement_df_group_zcta19.drop(columns=['year'])

law_enforcement_df_group_zcta19.head()
```

```python
#saving file
law_enforcement_df_group_zcta19.to_csv('law_enforcement_df_group_zcta19', index=False) 
```
